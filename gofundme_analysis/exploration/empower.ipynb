Rough prototype. Working with the dependency tree to break down how people use the metaphor.

```python
lst = labeled.loc[labeled['type'] == 'battle'].sample(20)

for ix, l in lst.iterrows():
    print(nlp(l['fragment']))
    print()
    for w in nlp(l['fragment']):
        
        if w.pos_ == 'VERB' and w.lemma_ in ['fight', 'battle', 'beat', 'defeat']:
            print('V', list(w.subtree))
        elif w.dep_ in ['dobj', 'pobj'] and w.text in ['fight', 'battle', 'war']:
            print('N', list(w.subtree))
        elif w.text in ['fight', 'battle', 'war']:
            print(list(w.subtree), w.dep_, w.tag_)
            
    print('_'*40)
```
